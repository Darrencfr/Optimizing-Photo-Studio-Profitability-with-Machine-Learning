# **Importing**

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

**Importing Dataset**

```python
df_studio_gambar = pd.read_csv('SALES_SYOKSENDIRISTUDIO.csv')
```

```python
df_studio_gambar
```

# **Feature Engineering**

```python
df_studio_gambar['Year'] = 2024

df_studio_gambar
```

```python
# Assigning Unique IDs to the Packages

df_studio_gambar['Package ID'] = pd.factorize(df_studio_gambar['Package'])[0]

df_studio_gambar
```

```python
# Assigning Unique Booking IDs to each reservation
bookings = []

booking_id = 1
while booking_id <= len(df_studio_gambar):
    bookings.append(booking_id)
    booking_id += 1

df_studio_gambar['Reservation ID'] = bookings

df_studio_gambar
```

```python
stopwords = ['Package', 'Pax']

def create_package_type(package):
    for st in stopwords:
        if st in package:
            if " - " in package:
                package_name, package_type = package.split(" - ")
                return pd.Series([package_name, package_type])
            else:
                return pd.Series([package_name, None])
    return pd.Series([package, None])

df_studio_gambar[['Package', 'Package Type']] = df_studio_gambar['Package'].apply(create_package_type)

df_studio_gambar
```

```python
month_mapping = {
    1 : 'January',
    2 : 'February',
    3 : 'March',
    4 : 'April',
    5 : 'May',
    6 : 'June',
    7 : 'July',
    8 : 'August',
    9 : 'September',
    10 : 'October',
    11 : 'November',
    12 : 'December'
}

valid_months = df_studio_gambar['Month'].isin(month_mapping.values())
df_studio_gambar['Month'] = df_studio_gambar['Month'].where(valid_months, df_studio_gambar['Month'].map(month_mapping))

df_studio_gambar
```

```python
month_mapping_reversi = {m: n for n, m, in month_mapping.items()}.copy()
print(month_mapping_reversi)
```

```python
df_studio_gambar['Cost Per Package (RM)'] = 0.9 * df_studio_gambar['Package Price (RM)']

df_studio_gambar
```

```python
df_studio_gambar['Total Profit (RM)'] = df_studio_gambar['Total Sale (RM)'] - (df_studio_gambar['Cost Per Package (RM)'] * df_studio_gambar['Total Quantity'])

df_studio_gambar
```

```python
# Add profit per reservation (RM)
df_studio_gambar['Profit Per Reservation (RM)'] = df_studio_gambar['Package Price (RM)'] - df_studio_gambar['Cost Per Package (RM)']

df_studio_gambar
```

```python
df_studio_gambar = df_studio_gambar[['Reservation ID', 'Year', 'Month', 'Package Type', 'Package ID', 'Package', 'Package Price (RM)', 'Total Quantity', 'Total Sale (RM)', 'Cost Per Package (RM)', 'Total Profit (RM)', 'Profit Per Reservation (RM)']]

df_studio_gambar
```

# **Model Building and Training**

```python
encoder = LabelEncoder()
scaler = StandardScaler()
```

```python
df_studio_gambar['Package Type'] = encoder.fit_transform(df_studio_gambar['Package Type'])

x = df_studio_gambar.drop(['Reservation ID', 'Year', 'Month', 'Package', 'Total Profit (RM)'], axis=1)
y = df_studio_gambar['Total Profit (RM)']
```

```python
x
```

```python
y
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

**Random Forest Regressor**

```python
RFR_model = RandomForestRegressor(n_estimators=100, random_state=42)
RFR_model.fit(x_train, y_train)
```

```python
y_pred_RFR = RFR_model.predict(x_test)
```

**Model Evaluation**

```python
print(f"R2 Score: {r2_score(y_test, y_pred_RFR) * 100:.2f}%")
print(f"MSE: {mean_squared_error(y_test, y_pred_RFR)}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_RFR)}")
```

**Visualising Feature Importances**

```python
feature_importances = np.abs(RFR_model.feature_importances_)
feature_names = x.columns

plt.figure(figsize=(10, 6))
plt.barh(range(len(feature_importances)), feature_importances, align='center', color='skyblue', edgecolor = 'black')
plt.yticks(np.arange(len(feature_importances)), feature_names)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importances in Random Forest Regression')
plt.show()
```

**Multiple Linear Regression**

```python
MLR_model = LinearRegression()
MLR_model.fit(x_train, y_train)
```

```python
y_pred_MLR = MLR_model.predict(x_test)
```

**Model Evaluation**

```python
print(f"R2 Score: {r2_score(y_test, y_pred_MLR) * 100:.2f}%")
print(f"MSE: {mean_squared_error(y_test, y_pred_MLR)}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_MLR)}")
```

**Visualising Coefficients**

```python
coefficients = np.abs(MLR_model.coef_)
feature_names = x.columns

plt.figure(figsize=(10, 6))
plt.barh(range(len(coefficients)), coefficients, align='center', color='skyblue', edgecolor = 'black')
plt.yticks(np.arange(len(coefficients)), feature_names)
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.title('Feature Importance in Multiple Linear Regression')
plt.show()
```
